Tax Calculator — Indian Income Tax (New Regime FY 2024-25)

```python
# Indian Income Tax Calculator (New Regime FY 2024-25)
# With Indian Number Formatting

def format_indian_currency(amount):
    """Format number in Indian numbering system."""
    amount = round(amount, 2)
    s = str(int(amount))
    decimal = f"{amount:.2f}".split('.')[1]

    if len(s) <= 3:
        return s + "." + decimal

    last_three = s[-3:]
    remaining = s[:-3]

    parts = []
    while len(remaining) > 2:
        parts.insert(0, remaining[-2:])
        remaining = remaining[:-2]

    if remaining:
        parts.insert(0, remaining)

    return ",".join(parts) + "," + last_three + "." + decimal


def calculate_tax(income):
    standard_deduction = 75000
    taxable_income = income - standard_deduction

    print("\n========== INCOME TAX CALCULATION ==========")
    print(f"Gross Income: ₹{format_indian_currency(income)}")
    print(f"Standard Deduction: ₹{format_indian_currency(standard_deduction)}")

    if taxable_income <= 0:
        print("No tax applicable after deduction.")
        return

    print(f"Taxable Income: ₹{format_indian_currency(taxable_income)}\n")

    slabs = [
        (300000, 0.00),
        (700000, 0.05),
        (1000000, 0.10),
        (1200000, 0.15),
        (1500000, 0.20),
        (float('inf'), 0.30)
    ]

    previous_limit = 0
    total_tax = 0

    print("--------- SLAB BREAKDOWN ---------")

    for limit, rate in slabs:
        if taxable_income > previous_limit:
            income_in_slab = min(taxable_income, limit) - previous_limit
            tax_for_slab = income_in_slab * rate
            total_tax += tax_for_slab

            print(f"₹{format_indian_currency(previous_limit)} "
                  f"to ₹{format_indian_currency(limit)}")
            print(f"   Income in slab : ₹{format_indian_currency(income_in_slab)}")
            print(f"   Tax Rate       : {int(rate*100)}%")
            print(f"   Tax in slab    : ₹{format_indian_currency(tax_for_slab)}\n")

            previous_limit = limit

    effective_rate = (total_tax / income) * 100

    print("===================================")
    print(f"Total Tax Payable: ₹{format_indian_currency(total_tax)}")
    print(f"Effective Tax Rate: {round(effective_rate,2)}%")
    print("===================================")


# Main
annual_income = float(input("Enter your Annual Income (₹): "))
calculate_tax(annual_income)
```
